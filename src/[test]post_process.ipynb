```python
import numpy as np
import matplotlib.pylab as plt
```

```python
x = np.random.rand(30)
x = x > 0.65
x = x.astype(float)
plt.plot(x)
plt.show()
```

```python
flag = False
cnt = 0
y = x.copy()
for idx in range(len(x)):
    if x[idx] == 1 :
        flag = True
        cnt = 2
    else :
        # smoothing
        if cnt != 0 :
            y[idx]=1
            cnt-=1
        else :
            flag = False
plt.plot(y)
plt.show()
```
